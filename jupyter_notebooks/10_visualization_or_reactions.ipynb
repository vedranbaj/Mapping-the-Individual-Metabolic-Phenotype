# Feature importance visualization

```python
# --- Configuration ---
input_file = r"../final_result/avg_importances_diet.csv"

# --- 1. Load Data ---
if os.path.exists(input_file):
    df_imp = pd.read_csv(input_file)
    print(f"Loaded {len(df_imp)} features from {input_file}")

if not df_imp.empty:
    # --- 2. Process Data ---
    # Sort by Importance (Descending) and take Top 20
    top_n = 20
    df_top = df_imp.sort_values("AvgImportance", ascending=False).head(top_n)

    # --- 3. Plotting ---
    plt.figure(figsize=(10, 6)) # Adjust size as needed
    
    # Create Bar Plot
    # orient='v' makes vertical bars (Feature on X, Importance on Y)
    sns.barplot(
        data=df_top,
        x="Feature",
        y="AvgImportance",
        color="skyblue",  # Simple color
        edgecolor="black" # Adds a border for clarity
    )

    # --- 4. Formatting ---
    plt.title(f"Top {top_n} Most Important Metabolic Reactions", fontsize=15, pad=20)
    plt.xlabel("Reaction ID", fontsize=12)
    plt.ylabel("Importance (Mean Decrease Impurity)", fontsize=12)
    
    # Rotate x-axis labels for readability
    plt.xticks(rotation=45, ha="right", fontsize=10)
    
    # Add gridlines for easier reading of values
    plt.grid(axis='y', linestyle='--', alpha=0.7)
    
    plt.tight_layout()
    plt.show()
```

# Visualizations of reactions

```python
sampling_data_100 = pd.read_csv("../100_combined_samples.csv") # load the flux matrix
```

```python
sns.set_style("white")

sampling_data_100["Individual"] = sampling_data_100["Unnamed: 0"].str.extract(r"_(FAC\d+)_")

top_reactions = [
    "MAR06298",
    "MAR01331",
    "MAR08741",
    "MAR01530",
    "MAR07875",
    "MAR03952",
]

reaction_titles = {
    "MAR06298": "MAR06298 - PEP [c] + cis-aconitate [m] ↔ PEP [m] + cis-aconitate [c]",
    "MAR01331": "MAR01331 - prostaglandin D2 [c] ↔ prostaglandin D2 [r]",
    "MAR08741": "MAR08741 - fumarate [c] + sulfite [m] ↔ fumarate [m] + sulfite [c]",
    "MAR01530": "MAR01530 - cholesterol [c] ↔ cholesterol [r]",
    "MAR07875": "MAR07875 - ADP [m] + UDP [m] ↔ ATP [m] + UMP [m]",
    "MAR03952": "MAR03952 - THF [c] ↔ THF [m]",
}

max_distributions = 275 # 20, limit to avoid overcrowding, this was ran with 20 and 275 for figures in the report

n = len(top_reactions)
ncols = 2
nrows = math.ceil(n / ncols)

fig, axes = plt.subplots(nrows, ncols, figsize=(18, 14))
axes = axes.flatten()

all_individuals = sampling_data_100["Individual"].dropna().unique()

for i, rxn in enumerate(top_reactions):
    ax = axes[i]
    plotted = 0

    for mouse_id in all_individuals:
        subset = sampling_data_100.loc[
            sampling_data_100["Individual"] == mouse_id, rxn
        ].dropna()

        # only count valid distributions for this specific reaction
        if len(subset) < 2 or subset.nunique() < 2:
            continue

        sns.kdeplot(
            x=subset,
            ax=ax,
            fill=True,
            alpha=0.25,
            linewidth=0.8,
            bw_adjust=0.7,
            warn_singular=False
        )

        plotted += 1
        if plotted >= max_distributions:
            break

    ax.set_yscale("symlog", linthresh=1e-3)
    ax.set_title(reaction_titles.get(rxn, rxn), fontsize=20)
    ax.set_xlabel("Flux (mmol/liver/day)", fontsize=16)
    ax.set_ylabel("log(Density)", fontsize=16)
    ax.tick_params(axis='y', labelsize=14)  # larger y-tick labels
    ax.grid(False)

for j in range(i + 1, len(axes)):
    axes[j].axis("off")

plt.tight_layout()
plt.show()
```

```python
def get_condition(data_with_conditions, row_name='Diet'):
    '''
    Extract diet information from the metadata file.
    
    Inputs:
        data_with_conditions : str
            Path to the metadata CSV file containing 'Diet' and 'OmicsEarTag' rows.
        row_name : str
            Name of the row to extract (e.g., 'Diet').
    
    Outputs:
        pd.DataFrame - Clean mapping of FAC_ID to Diet
    '''
    # Load metadata
    meta = pd.read_csv(data_with_conditions, index_col=0)
    
    # Extract relevant rows 
    diet_row = meta.loc[row_name]
    fac_row = meta.loc['OmicsEarTag']
    
    # Build mapping
    diet_map = pd.DataFrame({
        'FAC_ID': fac_row.values,
        row_name: diet_row.values
    })
    
    # Drop unwanted rows (header-like entries)
    diet_map = diet_map[~diet_map['FAC_ID'].isin(['Orig_ID', 'IDFemaleAgingColony'])].reset_index(drop=True)
    
    return diet_map
```

```python
diet_data = get_condition("../aData_S1_AllOmicsandPhenotypeData.csv", row_name='Diet')
```

```python
# Extract individual IDs from the sample names if not already done
sampling_data_100['Individual'] = sampling_data_100['Unnamed: 0'].str.extract(r'_(FAC\d+)_')

# --- CRITICAL STEP: Merge Diet info into the sampling data ---
# We merge on 'FAC_ID' (from diet_data) and 'Individual' (from sampling_data)
merged_data = sampling_data_100.merge(
    diet_data, 
    left_on='Individual', 
    right_on='FAC_ID', 
    how='left'
)
```

```python
sns.set_style("white")

# -----------------------------
# Configuration
# -----------------------------
diet_col = "Diet"

reactions = [
    "MAR04109",
    "MAR03793",
    "MAR03792",
    "MAR07809",
    "MAR04437",
    "MAR04656",
]

rxn_titles = {
    "MAR04109": "MAR04109 - AKG [m] + alanine [m] ↔ glutamate [m] + pyruvate [m]",
    "MAR03793": "MAR03793 - 2-methyl-3-oxopropanoate [m] + alanine [m]\n↔ D-3-amino-isobutanoate [m] + pyruvate [m]",
    "MAR03792": "MAR03792 - 2-methyl-3-oxopropanoate [m] + glutamate [m]\n↔ AKG [m] + D-3-amino-isobutanoate [m]",
    "MAR07809": "MAR07809 - D-aspartate [c] ↔ D-aspartate [e]",
    "MAR04437": "MAR04437 - histidine [c] ↔ H+ [c] + NH3 [c] + urocanate [c]",
    "MAR04656": "MAR04656 - NADPH [m] + folate [m]\n→ NADP+ [m] + dihydrofolate [m]",
}

diet_color_map = {
    "CD": "#ef6a61",
    "HF": "#5b6fff",
}

max_distributions = 275  # keep same overall density count behavior as your reference style

# -----------------------------
# Layout: same general sizing style as your reference
# -----------------------------
n = len(reactions)
ncols = 2
nrows = math.ceil(n / ncols)

fig, axes = plt.subplots(nrows, ncols, figsize=(18, 14))
axes = axes.flatten()

all_individuals = merged_data["Individual"].dropna().unique()

for i, rxn in enumerate(reactions):
    ax = axes[i]
    plotted = 0

    for mouse_id in all_individuals:
        subset = merged_data.loc[
            merged_data["Individual"] == mouse_id, [rxn, diet_col]
        ].dropna()

        # only plot valid distributions
        if subset.empty:
            continue
        if len(subset[rxn]) < 2 or subset[rxn].nunique() < 2:
            continue

        diet = subset[diet_col].iloc[0]
        color = diet_color_map.get(diet, "gray")

        sns.kdeplot(
            x=subset[rxn],
            ax=ax,
            color=color,
            fill=True,
            alpha=0.25,
            linewidth=0.8,
            bw_adjust=0.7,
            warn_singular=False,
            cut=0,
            legend=False,
        )

        plotted += 1
        if plotted >= max_distributions:
            break

    ax.set_yscale("symlog", linthresh=1e-3)
    ax.set_title(rxn_titles.get(rxn, rxn), fontsize=20)
    ax.set_xlabel("Flux (mmol/liver/day)", fontsize=16)
    ax.set_ylabel("log(Density)", fontsize=16)
    ax.tick_params(axis="x", labelsize=14)
    ax.tick_params(axis="y", labelsize=14)
    ax.grid(False)

# Hide unused axes
for j in range(i + 1, len(axes)):
    axes[j].axis("off")

# Shared legend on center right
legend_handles = [
    Patch(facecolor=diet_color_map[d], edgecolor=diet_color_map[d], alpha=0.25, label=d)
    for d in diet_color_map
]

fig.legend(
    handles=legend_handles,
    labels=[h.get_label() for h in legend_handles],
    title="Diet",
    loc="center right",
    frameon=True,
    fontsize=16,
    title_fontsize=18,
)

plt.tight_layout(rect=[0, 0, 0.90, 1])
plt.show()
```
